# Visualization

```python
# univariante analysis
plt.figure(figsize=(15,20))
sns.set_style('darkgrid')
plt.suptitle('Univariant Analysis')
for i,col in enumerate( df.columns):
   if df[col].dtype !='object':
      plt.subplot(3,3,i +1)
      plt.title(col)
      sns.kdeplot(df[col],color='skyblue')
plt.tight_layout()
plt.show()
		
   
```
